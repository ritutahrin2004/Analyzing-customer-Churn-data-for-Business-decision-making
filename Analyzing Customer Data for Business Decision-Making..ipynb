```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\user\Downloads\train.csv.zip")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
df = data.fillna(data.mean(numeric_only=True))
print(df)
```

```python
data.drop_duplicates(inplace=True)
print(data)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['Sales']] = scaler.fit_transform(data[['Sales']])
print(data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['Sales']] = scaler.fit_transform(data[['Sales']])
print(data)
```

```python
data = pd.get_dummies(data,columns=['Category'])
print(data)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['City'] = le.fit_transform(data['City'])
print(data)
```

```python
print(data.columns)
```

```python
data['Country'] = data['Country'].str.lower().str.strip()
print(data)
```

```python
data['Country'].replace({'wrong_value': 'correct_value'}, inplace = True)
print(data)
```

```python
pie_chart = data.groupby('Region')['Sales'].mean().reset_index()
```

```python
color = ['orange','gray', 'red', 'blue']
plt.pie(pie_chart['Sales'], labels = pie_chart['Region'], shadow = True, colors = color)
plt.title('Region Wise Sales')
plt.show
```
